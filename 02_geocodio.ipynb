```python
import pandas as pd
```

## Let's import the file Geocodio gave you

```python
geocoded = pd.read_csv('outputs/intersections_geocodio_c6044031aa6c588cb00ad09b9cd26d716986be34.csv', dtype={'Zip': object})
```

```python
geocoded.tail(3)
```

```python
geocoded = geocoded[[
    'intersection', 'Latitude', 'Longitude', 'Accuracy Score',
    'Accuracy Type', 'City', 'State', 'County', 'Zip',
]].copy()
geocoded
```

## Merge it back to the original dataset

```python
geocodable = pd.read_csv('outputs/geocodable.csv', parse_dates=['timestamp'])
```

```python
geocodable.tail(3)
```

```python
richmond_crashes_geocoded = pd.merge(geocodable, geocoded, how='left', on='intersection', validate='m:1')
```

```python
richmond_crashes_geocoded
```

## Export

```python
richmond_crashes_geocoded.to_csv('outputs/richmond_crashes_geocoded.csv', index=False)
```
